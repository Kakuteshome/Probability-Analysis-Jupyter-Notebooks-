#Q13

```python
# Probabilities
p_A = 0.2
p_B = 0.3

# Calculate complement probabilities
p_A_comp = 1 - p_A
p_B_comp = 1 - p_B

# Create Venn diagram
venn2(subsets=(p_A, p_B, 0), set_labels=(f'A\nP(A) = {p_A:.2f}', f'B\nP(B) = {p_B:.2f}'),
      set_colors=('skyblue', 'lightgreen'), alpha=0.7)

# Show the plot
plt.title('Venn Diagram with Probability Values')
plt.show()
```
